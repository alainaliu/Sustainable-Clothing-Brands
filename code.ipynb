```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

**HTTP REQUEST**

```python
website = "https://maylynandco.com/collections/lingerie"
response = requests.get(website) #get request
```

**STATUS CODE CHECK**

```python
response.status_code #see 200 successful http request
```

**SOUP OBJECT**

```python
soup = BeautifulSoup(response.content, "html.parser") #html parser
```

```python
soup #html of website
```

**Results**

```python
results = soup.find_all('li', {'class':'grid__item'})
```

```python
len(results)
```

```python
results[0] #first product
```

**Target Necessary Data**

```python
#Name First Product
results[0].find('a', {'class': 'full-unstyled-link'}).get_text().strip()
```

```python
#Price First Product
results[0].find('span', {'class': 'price-item price-item--regular'}).get_text().strip()
```

**Relative URL**

```python
#relative url
relative_url = results[0].find('a', {'class':'full-unstyled-link'}).get('href')
```

```python
#root element inside var
root_url = "https://maylynandco.com/"
```

```python
#combine root element + relative url
full_url = root_url + relative_url
full_url
```
