```python
item = []
price = []
relative_url = []
```

```python
for i in range(len(results)):
    item.append(results[i].find('a', {'class': 'full-unstyled-link'}).get_text().strip())
    price.append(float(results[i].find('span', {'class': 'price-item price-item--regular'}).get_text().strip().replace("$", "").replace(" USD", "").replace("From ", "")))
    relative_url.append(results[i].find('a', {'class':'full-unstyled-link'}).get('href'))
```

```python
item
price
relative_url
```

```python
len(full_url)
```

```python
maylynandco = {'item': item,
               'price': price,
               'relative_url': relative_url,
               'full_url': [root_url + i for i in relative_url]}
maylynandco
```

```python
maylynandco_df = pd.DataFrame(maylynandco)
maylynandco_df['store'] = "maylynandco"
maylynandco_df.set_index('store', inplace=True)
maylynandco_df
```
